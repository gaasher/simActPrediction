```python
import pandas as pd
import numpy as np
import os 
from tqdm import tqdm as tqdm
```

```python
# loop through all nih data files, and collect value counts of number of activities in 'Activity Label Provided By User' column
value_counts = {}

for file in tqdm(os.listdir('../data/nih/')):
    if file.endswith('.csv'):
        df = pd.read_csv('../data/nih/' + file)
        value_counts[file] = df['Activity Label Provided By User'].value_counts()

agg_value_counts = {}
# aggregate value counts for all files
for file in value_counts.keys():
    for activity in value_counts[file].index:
        if activity in agg_value_counts:
            agg_value_counts[activity] += value_counts[file][activity]
        else:
            agg_value_counts[activity] = value_counts[file][activity]
agg_value_counts
```

```python
# count the number of files each activity appears in
activity_files = {}
for file in value_counts.keys():
    for activity in value_counts[file].index:
        if activity in activity_files:
            activity_files[activity] += 1
        else:
            activity_files[activity] = 1
# sort the dictionary by the number of files each activity appears in
activity_files = dict(sorted(activity_files.items(), key=lambda item: item[1], reverse=True))
# sort agg_value_counts by the number of files each activity appears in
agg_value_counts = dict(sorted(agg_value_counts.items(), key=lambda item: item[1], reverse=True))

# combine these 2 dictionaries into a dataframe, impute 0 for missing values
df = pd.DataFrame(data=[agg_value_counts, activity_files]).T.fillna(0)
df.columns = ['count', 'files']

categories = {
    'work': ['Work', 'Computer Work', 'Websurf', 'Computer', 'Meeting'],
    'eat': ['Eat'],
    'leisure': ['Relax', 'Watch TV', 'Read', 'Napping ', 'Beach', 'Getting Hair Cut'],
    'active_leisure': ['Play', 'Garden', 'Shop', 'Shopping', 'Shopping '],
    'exercise': ['Exercise', 'Walk', 'Run', 'Cycle', 'Dog Walk'],
    'social': ['Socialize', 'Church', 'Church ', 'Pray ROSARY', 'Church Services'],
    'housework': ['Clean', 'Laundry', 'Laundry ', 'Take Medicine', 'Cook'],
    'personal_care': ['Dress', 'Bathe', 'Sleep', 'Dress'],
    'travel': ['Drive', 'Travel', 'Driving'],
    'other': ['Test',]
}

# get aggregate category counts for each category as well as the number of files each category appears in, 
category_counts = {}
category_files = {}
for file in value_counts.keys():
    for category in categories:
        added = False
        for activity in value_counts[file].index:
            if activity in categories[category]:
                if category in category_counts:
                    category_counts[category] += value_counts[file][activity]
                else:
                    category_counts[category] = value_counts[file][activity]

                if category in category_files and not added:
                    category_files[category] += 1
                    added = True
                elif not added:
                    category_files[category] = 1
                    added = True
category_counts = dict(sorted(category_counts.items(), key=lambda item: item[1], reverse=True))
category_files = dict(sorted(category_files.items(), key=lambda item: item[1], reverse=True))

# combine these 2 dictionaries into a dataframe, impute 0 for missing values
category_df = pd.DataFrame(data=[category_counts, category_files]).T.fillna(0)
category_df.columns = ['count', 'files']
category_df
```

```python
'''
Before doing any actions, we create a mapping of each activity label to a number, then we also map each category to a number
'''
category_mapping = {}
activity_mapping = {}
for i, category in enumerate(category_counts.keys()):
    category_mapping[category] = i
for i, activity in enumerate(agg_value_counts.keys()):
    activity_mapping[activity] = i
print(category_mapping)

'''
1) Load through every file in nih folder
2) For each file, load the file into a dataframe
3) make a new dataframe with rows that have a non-null 'Activity Label Provided By User' column
4) for each row in the new dataframe, do the following:
    a) if category of activity label is other, ignore the row
    b) log the sensor data ID in some sort of list so that we can prevent duplicates in dataset
    c) if the sensor data ID is not within 10 seconds of a sensor data ID in the list, continue
    d) using the original df take the 30 minute window around the activity label (15 minutes before and 15 minutes after), and add it to the new dataframe \\
       this is calculated by taking the sensor data ID and subtracting 15*60 from it, then adding 15*60 to it (each sensor data ID is 1 second apart) \\
    e) we also add a unique sample ID to each row which is added to the new dataframe (one id per activity label), and a user id
    f) add two columns to the new dataframe, one for the category of the activity label, and one for the activity label
4) return the new dataframe
'''

# Initialize an empty dataframe and list for sensor data IDs
new_df = pd.DataFrame()
sample_id = 0  # Initialize a sample ID


for file in tqdm(os.listdir('../data/nih/')):
    sensor_data_ids = []

    user_id = file[25:28]
    df = pd.read_csv('../data/nih/' + file)

    # Filter the dataframe to only include rows with non-null 'Activity Label Provided By User'
    filtered_df = df[df['Activity Label Provided By User'].notnull()]

    # Process each row in the filtered dataframe
    for index, row in tqdm(filtered_df.iterrows()):
        if row['Activity Label Provided By User'] == 'Test':
            continue
        
        sensor_data_id = row['Sensor Data ID']
        
        # Check if sensor data ID is already processed or within 10 seconds of an existing one
        if all(abs(sid - sensor_data_id) > 10 for sid in sensor_data_ids):
            sensor_data_ids.append(sensor_data_id)  # Add current sensor data ID to the list
            
            # Select rows within the 30-minute window around the current sensor data ID
            window_start = sensor_data_id - 15*60
            window_end = sensor_data_id + 15*60
            windowed_df = df[(df['Sensor Data ID'] >= window_start) & (df['Sensor Data ID'] <= window_end)].copy()
            
            # Assign a unique sample ID to each windowed data and concatenate it to the new dataframe
            windowed_df['sample_id'] = sample_id
            windowed_df['Activity'] = activity_mapping[row['Activity Label Provided By User']]
            # for category mapping we first find the category of the activity label
            category = ''
            for key in categories:
                if row['Activity Label Provided By User'] in categories[key]:
                    category = key
                    break
            windowed_df['Category'] = category_mapping[category]
            windowed_df['Category_label'] = category
            windowed_df['user_id'] = user_id

            # resort windowed_df by sensor data ID
            windowed_df.sort_values(by='Sensor Data ID', inplace=True)
            
            # if length of windowed_df is less than 64
            if len(windowed_df) < 64:
                continue
            else:
                new_df = pd.concat([new_df, windowed_df])
            
            sample_id += 1  # Increment the sample ID for the next window

new_df.to_csv('../data/processed_datasets/nih_processed_df.csv', index=False)
```

```python
'''
Next steps: actually divide the dataset into train, validation, and test sets (stratified by user_id and category), 64 rows per sample,
 with missing IDs imputed, and sliding window of 32 rows
'''

# Split the dataset into train, validation, and test sets with stratification on sample and user
from sklearn.model_selection import train_test_split

# get unique experiment ids
unique_user_ids = new_df['user_id'].unique()

# split the unique experiment ids into train, validation, and test sets
train_user_ids, test_user_ids = train_test_split(unique_user_ids, test_size=0.15, random_state=25)
train_user_ids, val_user_ids = train_test_split(train_user_ids, test_size=0.15, random_state=25)

train_df = new_df[new_df['user_id'].isin(train_user_ids)]
val_df = new_df[new_df['user_id'].isin(val_user_ids)]
test_df = new_df[new_df['user_id'].isin(test_user_ids)]

# get value counts for each category in each dataset
train_category_counts = train_df['Category_label'].value_counts(normalize=True)
val_category_counts = val_df['Category_label'].value_counts(normalize=True)
test_category_counts = test_df['Category_label'].value_counts(normalize=True)

# assert that the no category has more than 2% difference in frequency between the datasets
assert(max(abs(train_category_counts - val_category_counts)) < 0.10 and max(abs(train_category_counts - test_category_counts)) < 0.10 and max(abs(val_category_counts - test_category_counts)) < 0.10)    
      

print(train_category_counts, val_category_counts, test_category_counts)
```

```python
'''
Next we create 3 numpy arrays for our train, validation, and test sets, and save them to disk, each with 3 dimensions: (num_samples, 64, num_features), 
with a sliding window of 32 rows (32s)
'''

features_list=['Yaw (rad)', 'Pitch (rad)','Roll (rad)','Rotation Rate X (rad/s)','Rotation Rate Y (rad/s)','Rotation Rate Z (rad/s)',
               'User Acceleration X (m/s^2)','User Acceleration Y (m/s^2)','User Acceleration Z (m/s^2)','Altitude (m)','Horizontal Accuracy (m)',
               'Speed (m/s)']

print(f'Number of features: {len(features_list)}')

# drop any rows with missing values in features_list
train_df = train_df.dropna(subset=features_list)
val_df = val_df.dropna(subset=features_list)
test_df = test_df.dropna(subset=features_list)

def create_dataset(df, features_list):
    # Initialize a list to hold the data windows
    data_windows = []
    identifiers = ['Activity', 'user_id', 'sample_id', 'Category']  # Identifiers to be added

    for sample_id in tqdm(df['sample_id'].unique()):
        sample_df = df[df['sample_id'] == sample_id]
        for start in range(0, len(sample_df), 16):  # Use a step of 16 for sliding window
            end = start + 64
            if end <= len(sample_df):
                window = sample_df.iloc[start:end][features_list].to_numpy()
                # Add identifiers only once per window
                id_values = sample_df.iloc[start][identifiers].to_numpy().reshape(1, -1)
                # Repeat the identifier values to match the window shape
                id_values = np.repeat(id_values, 64, axis=0)
                window = np.hstack((window, id_values))
                data_windows.append(window)
            else:
                # Handle padding if the window is less than 64
                window = sample_df.iloc[start:][features_list].to_numpy()
                pad_length = 64 - window.shape[0]
                # Pad the data window
                window_padded = np.pad(window, ((0, pad_length), (0, 0)), 'edge')
                id_values = sample_df.iloc[start][identifiers].to_numpy().reshape(1, -1)
                id_values = np.repeat(id_values, 64, axis=0)
                window_padded = np.hstack((window_padded, id_values))
                data_windows.append(window_padded)

    # Convert the list of windows into a numpy array
    dataset = np.array(data_windows)
    return dataset

# Create the datasets
train_npy = create_dataset(train_df, features_list)
val_npy = create_dataset(val_df, features_list)
test_npy = create_dataset(test_df, features_list)

print(f'Train dataset shape: {train_npy.shape}, Validation dataset shape: {val_npy.shape}, Test dataset shape: {test_npy.shape}')

# Save the datasets to disk
np.save('../data/processed_datasets/nih_train.npy', train_npy)
np.save('../data/processed_datasets/nih_val.npy', val_npy)
np.save('../data/processed_datasets/nih_test.npy', test_npy)
```
